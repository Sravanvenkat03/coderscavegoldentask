```python
df['diagnosis'].replace(["M","B"],[1,2],inplace=True)
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
print(x,y)
```

```python
import seaborn as sns
x=pd.DataFrame(df.drop('id',axis=1))
y=pd.DataFrame(df['id'])
for i in x:
    sns.histplot(df[i],label=i)
plt.show()
```

```python
print(df.columns)
sns.heatmap(df[df.columns],cmap = "tab20")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
print(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
y_pred=dtc.predict(x_train)
print(accuracy_score(y_train,y_pred)*100)
```

```python
y_test_pred=dtc.predict(x_test)
print(accuracy_score(y_test,y_test_pred)*100)
```
